```python
from bs4 import BeautifulSoup
import requests

url = "https://www.kitapsepeti.com/roman"  #The URL of the web page you want to parse

yanit = requests.get(url)
sayfa_icerigi = yanit.content
print(sayfa_icerigi)
```

```python
soup = BeautifulSoup(sayfa_icerigi, "html.parser")

# Selecting a single member
baslik = soup.find("h1")
print(baslik)
```

```python
# Staff selection by class
div_eleman = soup.find_all("div", class_="col-12")
print(div_eleman)
```

```python
#Staff selection by Id
id_eleman = soup.find(id="product-list-panel")
print(div_eleman)
```

```python

#find and print titles
title = soup.find('header')
print('Balklar:', title.text)
```
